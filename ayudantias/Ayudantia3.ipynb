Considere la siguiente ecuacin:

\begin{equation}
    f(x,y) = x + y \sin \left( \log (y^2 + 1) - y \right) = 1,
\end{equation}

Despejamos para que quede igualada a 0 y poder hacer una busqueda de raiz:

\begin{equation}
    F(x,y) = x + y \sin \left( \log(y^2 + 1) - y \right) - 1 = 0.
\end{equation}


Dado un valor de $x = \hat{x}$, debemos encontrar $y$ tal que:

\begin{equation}
    F(\hat{x}, y) = 0.
\end{equation}

El mtodo de Newton se define como:

\begin{equation}
    x_{n+1} = x_n - \frac{f( x_n)}{f'(x_n)}.
\end{equation}

Donde:
- $f(x) = F(\hat{x}, y)$.
- $f'(x) = \frac{d}{dy} F(\hat{x}, y)$.

Derivamos $F(y)$:

\begin{equation}
    F'(\hat{x}, y) = \sin \left( \log(y^2 + 1) - y \right) + y \cos \left( \log(y^2 + 1) - y \right) \left( \frac{2y}{y^2 + 1} - 1 \right).
\end{equation}

Finalmente:
\begin{equation}
    y_{n+1} = y_n - \frac{F(\hat{x}, y)}{F'(\hat{x}, y)}.
\end{equation}

```python
import numpy as np
f = lambda x,y: x + y * np.sin(np.log(y**2 + 1) - y)
F= lambda x,y: x + y * np.sin(np.log(y**2 + 1) - y) - 1
Fp= lambda x,y: np.sin(np.log(y**2 + 1) - y) + y*(np.cos(np.log(y**2 + 1) - y) * ((2 * y) / (y**2 + 1) - 1))

def newton_method(y0, x=2):
    y = y0
    for i in range(10):
        y -= F(x,y) / Fp(x,y)
        print(y)
    return y

# Ejecucin con valores iniciales
y0 = -6
y_result = newton_method(y0)
f(2., y_result)
```
